# Matplotlib Intro

## Two APIs
- matplotlib.pyplot
- object-oriented api

- matplotlib.pyplot
    - created to mimic MATLAB's plotting functionality
    - imperative or state-based: commands change the state of most recent figure
    - intended for simple cases or interactive plots

```python
# example
import numpy as np
import matplotlib.pyplot as plt

y = np.arange(10)
plt.figure(figsize=(9,7), dpi=100)
plt.plot(y, "bo-")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```

- object-oriented api
    - matplotlib is written in an object-oriented manner
    - provides more control and customization
    - typically create a figure or axes with high-level pyplot commands then use those object's methods to build the plot(s)

```python
# example
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

axes[0].plot(y, "bo-")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].grid(True)
axes[0].set_title("Plot 1")

y3 = y**3
axes[1].plot(y3, "bo-")
axes[1].set_xlabel("x")
axes[1].set_ylabel("$y^3$")
axes[1].grid(True)
axes[1].set_title("Plot 2")

_ = fig.suptitle("y and $y^3$")
```

## Figure Objects
- `Figure` 
    - the whole figure
    - contains child `Axes`
- `Axes` 
    - each plot in the figure
    - holds data in the plot, X and Y axis labels, plot title, etc
- `Axis`
    - number-line-like objects (X-axis and Y-axis)
    - responsible for setting plot limits and creating ticks and ticklabels
    - a `Locator` object determines the tick locations and a `Formatter` determines the ticklabel strings
- `Artists`
    - all elements rendered on figure (blue labels)
    - most are tied to an `Axes`
